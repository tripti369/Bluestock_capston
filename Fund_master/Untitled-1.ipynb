```python
Q1 = df['expense_ratio_pct'].quantile(0.25)
Q3 = df['expense_ratio_pct'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = df[
    (df['expense_ratio_pct'] < lower_bound) |
    (df['expense_ratio_pct'] > upper_bound)
]

print("Number of outliers:", len(outliers))
```

```python
outliers[['scheme_name', 'category', 'expense_ratio_pct']]
```

```python
for cat in df['category'].unique():
    temp = df[df['category'] == cat]

    Q1 = temp['expense_ratio_pct'].quantile(0.25)
    Q3 = temp['expense_ratio_pct'].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = temp[
        (temp['expense_ratio_pct'] < lower) |
        (temp['expense_ratio_pct'] > upper)
    ]

    print(f"{cat}: {len(outliers)} outliers")
```

```python
equity = df[df['category'] == 'Equity']

Q1 = equity['expense_ratio_pct'].quantile(0.25)
Q3 = equity['expense_ratio_pct'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

equity_outliers = equity[
    (equity['expense_ratio_pct'] < lower) |
    (equity['expense_ratio_pct'] > upper)
]

equity_outliers[['scheme_name', 'expense_ratio_pct']]
```
